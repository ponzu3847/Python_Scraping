# ２．　Wgetで始めるクローラー
![](./images/PastedGraphic_0.png)

### 基本的な使い方  
wget ダウンロードしたいファイルのURL  
wget -O ./Downloads/ファイル名 ダウロードしたいファイルのURL  
（カレントディレクトリにあるDownloadsフォルダにURLのファイルを保存）

```python
!wget -O ./Downloads/shoeisha.html https://www.shoeisha.co.jp
```

### ファイルに保存せずにダウンロードする（パイプ「｜」で他のコマンドに渡せる）
wget -O - ダウンロードしたいファイルのURL

```python
!wget -O - https://www.shoeisha.co.jp
```

### 再帰的にダウンロードする（再帰的に＝リンク先も探索する）  
wget -r -l1 ダウンロードしたいファイルのURL  
（1階層の深さまで再帰的にダウンロードする）

```python
!wget -r -l1 https://www.shoeisha.co.jp/book
```

### ダウンロード間隔を指定する
wget -r -l1 -w3 ダウンロードしたいファイルのURL  
（1階層の深さまで再帰的に3秒間隔でダウンロードする）

```python
!wget -r -l1 -w3 https://www.shoeisha.co.jp/book/
```

### 特定の拡張子のファイルだけダウンロードする
wget -r -l1 -A jpg,png,gif ダウンロードしたいファイルのURL  
（1階層の深さまで再帰的にjpg,png,gifのみダウンロードする）


```python
!wget -r -l1 -A jpg,png,gif https://www.shoeisha.co.jp/
```

### 特定の拡張子のファイルを除いてダウンロードする
wget -r -l1 -R jpg,png,gif ダウンロードしたいファイルのURL  
（1階層の深さまで再帰的にjpg,png,gifを除いてダウンロードする）

```python
!wget -r -l1 -R jpg,png,gif https://www.shoeisha.co.jp/
```

### 親ディレクトリをクローリング対象から外す
wget -r -np ダウンロードしたいファイルのURL

```python
!wget -r -np https://www.shoeisha.co.jp/book/
```

### 実際にクローリングしてみる

```python
!wget -r -l1 -w3 -np -R jpg,png,gif https://www.shoeisha.co.jp/book/
```

### teeコマンドでツリー表示をする

```python
!tree www.shoeisha.co.jp/
```

# ３．　UNIXコマンドでスクレイピング

## スクレイピングに役立つコマンド
    

### grep  
正規表現によるテキストの検索を行う  
grep パターン ファイル名

```python
!ps x | grep /usr/libexec
```

### sed
テキストの検索や置換を行う  
sed オプション コマンド ファイル名  

![](./images/PastedGraphic_2.png)

```python
#指定したテキストファイルから指定した文字列を検索して置換したテキストを返す
!sed s/android/Android/ sample.txt
```

```python
#指定したテキストファイルから指定した文字列を検索して置換し、新しいファイルとして保存する
!sed s/android/Android/ sample.txt > new_sample.txt
!cat new_sample.txt
```

### 正規表現の種類
* single regular expression（SRE、廃止が予定されている）
* 基本正規表現（BRE: basic regular expression）
* 拡張正規表現（ERE: Extended regular expression、-Eオプションを指定することで使えるようになる）
